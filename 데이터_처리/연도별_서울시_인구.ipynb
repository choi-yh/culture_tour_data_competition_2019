```python
pop_per_year = pd.concat([pop_2011,pop_2012,pop_2013,pop_2014,pop_2015,pop_2016,pop_2017,pop_2018],sort=False)
pop_per_year
```

```python
a = pop_per_year.iloc[:,[0,5,6,7]]
a
```

```
import pandas as pd
import numpy as np
import tensorflow as tf
```

```
data = pd.read_csv('/gdrive/My Drive/문화관광공모전/데이터_처리/tt.csv', encoding='utf-8')
data = data.iloc[:, 1:]
data.head()
```

```
from sklearn import preprocessing

df = pd.DataFrame(data.iloc[:, 2:])
df

min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(df)
df_normalized = pd.DataFrame(np_scaled)
df_normalized
```

```
norm_data = pd.concat([data.iloc[:, :2], df_normalized], axis=1)
norm_data
```

```
x_data = np.array(norm_data.iloc[:, 2:]).reshape(25,6)
y_data = np.array(norm_data.iloc[:, 1]).reshape(25,1)
```

```
x_data.shape, y_data.shape
```

```
tf.reset_default_graph()

X = tf.placeholder(tf.float32, shape=[None, 6])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W1 = tf.get_variable('W1', shape=[6, 10],
                    initializer=tf.contrib.layers.xavier_initializer())
b1 = tf.Variable(tf.random.normal([10]), name='bias1')
L1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.get_variable('W2', shape=[10, 1],
                    initializer=tf.contrib.layers.xavier_initializer())
b2 = tf.Variable(tf.random.normal([1]), name='bias2')

hypothesis = tf.sigmoid(tf.matmul(L1, W2) + b2)
cost = tf.reduce_mean(-(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis)))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

prediction = tf.cast(hypothesis > 0.5, tf.float32)
correct = tf.equal(prediction, Y)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

with tf.Session() as sess:
    sess.run(tf.initializers.global_variables())
    
    for step in range(10001):
        _, c = sess.run([optimizer, cost], feed_dict={X: x_data, Y: y_data})
        if step % 200 == 0:
            print("Step: {:04d}\tCost: {:.7f}".format(step, c))
            
    p, c, a = sess.run([prediction, correct, accuracy], feed_dict={X: x_data, Y: y_data})
    print("Accuracy: {:.2%}".format(a))
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression as LR
```

```
data = pd.read_csv("/gdrive/My Drive/문화관광공모전/데이터_처리/seoul_business (1).csv", encoding='utf-8', na_values=' ')
data = data.fillna(0)
data
```

```
region = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', 
          '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

print(len(region))
```

```
time = np.arange(0, 7).reshape(-1, 1)
rate = []

for i in range(len(region)):
    lr = LR()
    
    plt.figure()
    
    tmp = data[data['자치구'] == region[i]]
    avg = tmp['사업체 창업률']
    avg.fillna(0)
    
    plt.plot(time, avg, 'b')
    lr = lr.fit(time, avg)
    rate.append(lr.coef_[0])
    
    plt.xlabel('time')
    plt.ylabel('average')
    
    plt.show()
```

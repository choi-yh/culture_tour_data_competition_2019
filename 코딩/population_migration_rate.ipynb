```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```
data = pd.read_csv("/gdrive/My Drive/문화관광공모전/데이터_처리/population_migration.csv", encoding='utf-8')
data.head()
```

```
region = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', 
          '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

print(len(region))
```

```
for reg in region:
    plt.figure()
    tmp = data[data['자치구'] == reg]
    time = tmp['기간']
    avg = tmp['총전출']
    
    plt.plot(time, avg, 'b')
    
    plt.xlabel(reg)
    plt.ylabel('out')
    
    plt.show()
```

```
for reg in region:
    plt.figure()
    tmp = data[data['자치구'] == reg]
    time = tmp['기간']
    avg = tmp['총전입']
    
    plt.plot(time, avg, 'b')
    
    plt.xlabel(reg)
    plt.ylabel('out')
    
    plt.show()
```

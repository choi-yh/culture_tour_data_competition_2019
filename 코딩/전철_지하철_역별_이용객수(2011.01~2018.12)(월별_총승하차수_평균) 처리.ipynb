```
import pandas as pd
import numpy as np
```

```
tmp = pd.read_csv('/gdrive/My Drive/문화관광공모전/data/유동인구/전철_지하철_역별_이용객수(2011.01~2018.12)(월별_총승하차수_평균)/csv/강남구.csv', na_values='-')
tmp.head()
```

```
# 불필요한 부분 자르기

col = list(tmp.iloc[2])
data = tmp.iloc[3:, :]
data.columns = col
data.head()
```

```
# 전체 요일만 추려서 연도별로 바꿔주기
# 값들 숫자로 바꿔줘야함

data_col = data[['월', '모든요일']]
data_col.columns = ['월', '2011년', '2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년']
data_col.head()
```

```
# 데이터 타입 바꿔주기

col = list(data_col.columns)
series = [data_col['월']]
for year in col:
    if year != '월':
        series.append(pd.to_numeric(data_col[year]))
        
num_data = pd.concat([series[0], series[1]], axis=1)
for i in range(2, len(series)):
    num_data = pd.concat([num_data, series[i]], axis=1)
num_data.head()
```

```
# 월별로 평균내기

int(num_data[num_data['월']=='1월']['2018년'].mean())
```

```
months = list(num_data['월'].unique())
num_list = list(range(1, 13))
years = ['2011년', '2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년']
region = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

print(months, years, region)
print(len(months), len(years), len(region))
```

```
num_list
```

```
date = []
district = []
avg_people = []

for year in years:
    for i in range(len(months)):
        avg = int(num_data[num_data['월']==str(num_list[i])+'월'][year].mean()) # 해당 년도, 월의 평균 지하철 이용 수
        reg = region[0]
        
        # 각 리스트에 순차적으로 넣고 데이터 프레임을 만든다.
        date.append("{}{:02d}월".format(year, num_list[i]))
        district.append(reg)
        avg_people.append(avg)
```

```
final_data = pd.DataFrame(data={"기간": date, "자치구": district, "총 승하차 수": avg_people}, columns=["기간", "자치구", "총 승하차 수"])
final_data
```

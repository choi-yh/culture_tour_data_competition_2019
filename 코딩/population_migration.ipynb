```
import pandas as pd
```

```
data = pd.read_csv('/gdrive/My Drive/문화관광공모전/data/인구이동/서울시_인구이동(구별_월별)_통계(2014.01_2019.05)_1.csv', encoding='utf-8')
data = data.iloc[:1563, :]
data.head(), data.tail(), data.shape
```

```
len(data['기간'][3:]), data['기간'][3]
```

```
# 날짜 정리

for idx in range(3, len(data['기간'])):
    _tmp = data['기간'][idx].split('.')
    data['기간'][idx] = str(_tmp[0] + "년" + _tmp[1] + "월")

data['기간']
```

```
data_in = data.iloc[:, :3]
data_in.head()
```

```
# data_out = pd.concat([data.iloc[:, :2], data.iloc[:, 5]], axis=1)
data_out = data.iloc[:, 5]
data_out.head()
```

```
# 총전입, 총전출만 추리기

re_data = pd.concat([data_in, data_out], axis=1)
re_data = pd.DataFrame.drop(re_data, labels=[0,1,2], axis=0)
re_data = pd.DataFrame.rename(re_data, columns={'총이동': '총전입', '총이동.3': '총전출'})
re_data
```

```
region = list(set(re_data['자치구']))
region, len(region)
```

```
region = list(set(re_data['자치구'])) # 지역 분류
final_data = re_data[re_data['자치구'] == '종로구'] # 시작 데이터
region.remove('종로구') # 입력했으니 삭제

for reg in region:
    if reg != '서울특별시':
        _tmp = re_data[re_data['자치구'] == reg]
        final_data = pd.merge(final_data, _tmp, how='outer')
        
final_data # 11~13 년도를 따로 구해서 저장해야함
```

```
final_data.to_csv('/gdrive/My Drive/문화관광공모전/population_migration.csv', mode= 'w', sep=',')
```

```
import pandas as pd
```

```
# 데이터 정리

def handling(data):
    # 불필요한 부분 자르기
    col1 = list(data.iloc[2])
    pre_data = data.iloc[3:, :]
    pre_data.columns = col1
    
    # 전체 요일만 추려서 연도로 바꿔주기
    re_data = pre_data[['월', '모든요일']]
    re_data.columns = ['월', '2011년', '2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년']
    
    # 데이터 타입 변경
    col2 = list(re_data.columns)
    series = [re_data['월']] # pd.to_numeric을 위한 시리즈 분리
    for year in col2: # 분리한 시리즈 들의 데이터 타입 변경
        if year != '월':
            series.append(pd.to_numeric(re_data[year]))
    
    # 분리한 시리즈들 다시 합치기
    num_data = pd.concat([series[0], series[1]], axis=1)
    for i in range(2, len(series)):
        num_data = pd.concat([num_data, series[i]], axis=1)
    
    return num_data
```

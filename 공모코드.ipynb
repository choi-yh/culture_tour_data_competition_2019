```python
import pandas as pd
#판다스로 엑셀불러옴(데이터원본파일)
df = pd.read_excel('C:\\sample\\visit.xls', sheet_name='Sheet1')
df = df.replace('-', 0)
```

```python
#외국인통계를 구하는게 목적이니 조건검색
fdf= df[df.iloc[:,5]=='외국인']
fdf
```

```python
tt = fdf[fdf['군구'] == '종로구'].iloc[:,7:] # 종로구 날짜별(열수) 외국인 관광객수 7개의 관광지점(행수)
tt
```

```python
for col in tt.columns:
    tt[col] = tt[col].str.replace(",","") #위의 데이터 숫자들이 다 str형이기때문에 문자열,를 공백으로 바꿔줌
    
tt= tt.fillna(0) #그러면 원래의 0은 결측값으로 변하는데 그걸 다시 숫자 0으로 바꿔줌
for col in tt.columns:
    tt[col] = pd.to_numeric(tt[col]) #문자열 숫자들을 다 float형으로 바꿔줌
```

```python
ttsum = tt.sum(axis=0) # 열(날씨)별 합계
```

```python
df1 = pd.DataFrame({'시간':tt.columns,'지역': ['종로구' for x in range(len(tt.columns))],'외국인관광객수':ttsum[:]}) #다시 데이터프레임으로 만들어줌
```

```python
df1 # 여기까지가 예시
```

```python
#위의 작업들을 반복문으로 돌림
region = list(set(fdf['군구']))
region.remove('종로구') #위에서 이미 했기 때문에
print(region) 
```

```python
for reg in region:
    gg = fdf[fdf['군구'] == reg].iloc[:,7:]
    for col in gg.columns:
        gg[col] = gg[col].str.replace(",","")
    
    gg = gg.fillna(0)
    for col in gg.columns:
        gg[col] = pd.to_numeric(gg[col])
    ggsum = gg.sum(axis=0) #여기까진 위와 같고
    ggdf = pd.DataFrame({'시간': gg.columns,'지역' : [reg for x in range(len(gg.columns))], '외국인관광객수' : ggsum[:]})
    df1 = pd.concat([df1, ggdf], ignore_index=True) #처음에 만들어둔 데이터프레임과 병합, key는 시간으로 
    gg = '' #gg 초기화
    ggdf=''
```

```python
x_data = np.array(foreign.iloc[:,2:]).reshape(25,6)
y_data = np.array(foreign.iloc[:,1]).reshape(25,1)
```

```python
import tensorflow as tf

X = tf.placeholder(tf.float32, shape = [None, 6])
Y = tf.placeholder(tf.float32, shape = [None, 1])

W1 = tf.Variable(tf.random_normal([6, 1]), name='weight1') 
#열 값을 증가시키면 network가 wide해짐
b1 = tf.Variable(tf.random_normal([1]), name='bias1')

hypothesis = tf.sigmoid(tf.matmul(X, W1) + b1)

cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X : x_data, Y : y_data})
        if step % 1000 == 0 :
            print(step, cost_val)
            
    h, c, a ,w = sess.run([hypothesis, predicted, accuracy, W1], feed_dict={X:x_data, Y:y_data})
    print("\nhypothesis: ", h, "\nCorrect :", c , "\nAccuracy: ", a, "\nWeight :", w)
```
